```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_excel('result/result.xls')
```

```python
heights = np.array(df['GRIDCODE'])
visibility = np.array(df['visibility'])
n = int(heights.shape[0]**0.5)
heights = heights.reshape(n,n)
visibility = visibility.reshape(n,n)
plt.figure(figsize=(14,10))
sns.heatmap(heights, alpha=0.4);
sns.heatmap(visibility, alpha=0.4);
plt.plot(25, 25, color='green', marker='*', linestyle='dashed', linewidth=2, markersize=12);
```
